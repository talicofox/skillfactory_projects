В задании требуется сделать парсинг, вопросов и ответов викторины сайта с использованием selenium'а.

Несмотря на то, что html уже содержит все вопросы и варианты ответов и даже верные ответы особенность selenium в том,что получить можно только те элементы, которые изображаются на экране.

Поэтому придется пройти все 12 вопросов викторины и собрать варианты ответов  и верные ответы. 

Для сохранения результатов будут использованы 3 списка, которые возвращаются функцией.

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver_path = r'C:\Program Files (x86)\Google\Chrome\chromedriver.exe'

def run(driver_path):
    
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_experimental_option("prefs", {"profile.default_content_setting_values.cookies": 2}) 

    driver = webdriver.Chrome(driver_path,chrome_options=chrome_options)
    url = "https://tproger.ru/quiz/real-programmer/"
    driver.get(url)
    
    #define lists of questions, answers, right_answers
    list_questions = []
    list_answers = []
    list_right_answers = []
    
    button = driver.find_element_by_class_name("quiz_button")
    button.click()
    
    questions = driver.find_elements_by_class_name("question")
    quantity_questions = len(questions)
    
    id0 = 255
    
    for i in range(quantity_questions):
    
        print("\nQuestion ", i+1)
        item = driver.find_element_by_id('quiz_item_%i'%(id0+i,))
        
        question = item.find_element_by_class_name("question")
        print(question.text)
        list_questions.append(question.text)
                                                
        answer = item.find_element_by_xpath('//*[@id="quiz_item_%i"]/div[2]/div[1]/div'%(id0+i,)) 
        answer.click()
    
        answers = item.find_elements_by_class_name("answer_text")
        
        tmp = []
        for ans in answers:
            if ans.text:
                print(ans.text)
                tmp.append(ans.text)
        list_answers.append(tmp)
        
        
        right_answer = item.find_element_by_class_name("correct_answer").find_element_by_class_name("answer_text").text
        print("Right_answer - ", right_answer)
        list_right_answers.append(right_answer)
        
                
        #webdriver.ActionChains(driver).send_keys(Keys.SHIFT+Keys.PAGE_DOWN).perform()
        
        button = driver.find_element_by_xpath('//*[@id="quiz_item_%i"]/button'%(id0+i,))
        
        webdriver.ActionChains(driver).move_to_element(button).perform()
        button.click()
    
    driver.close()
    driver.quit()
    
    return list_questions, list_answers, list_right_answers


list_questions, list_answers, list_right_answers = run(driver_path)    
```
